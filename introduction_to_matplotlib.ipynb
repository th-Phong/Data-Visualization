### Python API vs Object-Oriented API
* Pyplot API -> Quickly
* OO API -> Advanced

### Pyplot API

```python
x  = np.linspace(1, 10, 1000)
plt.plot(x, np.sin(x), color="blue", linestyle='solid', label='Sine Wave');
plt.plot(x, np.cos(x), color="red", linestyle='dashed', label='Cosine Wave');

plt.title('Sine and Cosine Waves');
plt.xlabel('X-axis');
plt.ylabel('Y-axis');

# plt.xlim(0, 10);
# plt.ylim(-1.5, 1.5);
plt.axis([0, 10, -2, 1.5]);
# plt.axis("tight");
plt.axis("equal");

plt.legend();
plt.show();
```

### Object-Oriented API

```python
x = [1,2,3,4,5]
y = [11,12,13,14,15]

fig,ax = plt.subplots(figsize=(10,5))

ax.plot(x, y, color='purple', marker='o', linestyle='solid');
# ax.set_title('Sample Plot using Object-Oriented API');
# ax.set_xlabel('X-axis Label');
# ax.set_ylabel('Y-axis Label');
ax.set(
    title='Sample Plot using Object-Oriented API',
    xlabel='X-axis Label',
    ylabel='Y-axis Label'
);
```

## 2. Most common types of Matplotlib plots
* line
* scatter
* bar
* hist
* subplots()

### Line

```python
x = np.linspace(0, 10, 100)
x[:5]
```

```python
fig,ax = plt.subplots()
ax.plot(x, x**3, label='Cubic Function');
ax.set(
    title='Cubic Function Plot',
    xlabel='X-axis',
    ylabel='Y-axis'
);
ax.legend();
```

### Scatter

```python
plt.scatter(x, np.exp(x), label='Exponential Function', color='blue', linestyle='solid', alpha=0.7);
plt.title('Scatter Plot of Exponential Function');
plt.xlabel('X-axis');
plt.ylabel('Y-axis');
plt.legend();
plt.show();
```

```python
## OO API
rnd = np.random.RandomState(0);
x = rnd.randn(100);
y = rnd.randn(100);

fig,ax = plt.subplots();

colors = rnd.rand(100);
sizes = 1000 * rnd.rand(100);

scatter1 = ax.scatter(x, y, s=sizes, c=colors, alpha=0.5);
fig.colorbar(scatter1);

ax.set(
    title='Random Scatter Plot',
    xlabel='X-axis',
    ylabel='Y-axis',
);
plt.show();
```

### 3. Bar
* Vertical
* Horizontal

```python
#Prepare data from a python dictionary
soft_drink_prices = {
    "Coke": 10,
    "Pepsi": 20,
    "Sting": 15,
    "Sprite": 17,
    "Fanta": 12,
}

colors = ['red', 'blue', 'green', 'yellow', 'orange']

fig,ax = plt.subplots();
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values(), color=colors);

ax.set(
    title='Soft Drink Prices',
    xlabel='Soft Drink',
    ylabel='Price in $'
);

plt.show();
```

```python
fig,ax = plt.subplots();
ax.barh(soft_drink_prices.keys(), soft_drink_prices.values(), color=colors);
```

### 4. Histogram

```python
np.random.seed(50)
student_heights = np.random.normal(170, 10, 220)
student_heights[:20]
```

```python
plt.hist(student_heights, bins=15);
plt.title('Distribution of Student Heights');
plt.xlabel('Height (cm)');
plt.ylabel('Number of Students');
plt.show();
```
