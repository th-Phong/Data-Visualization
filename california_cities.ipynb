## Project: Data Visualization - California City

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('data/california_cities.csv')
print(cities.dtypes)
print("---------------------------------")
cities.head()
```

```python
# Extract Latitude and Longitude
latitudes = cities['latd'];
longitudes = cities['longd'];
population = cities['population_total'];
area = cities['area_total_km2'];

# print(latitudes.head());
# print(longitudes.head());

# rnd = np.random.RandomState(42)
# colors = rnd.rand(len(cities))
# print(plt.style.available)
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 6));

# plt.scatter(longitudes, latitudes, c=colors);
plt.scatter(
    longitudes, 
    latitudes, 
    c=np.log10(population), 
    s=area,
    cmap='viridis',
    alpha=0.7,
    linewidths=0);

plt.title('California Cities: Population and Area');
plt.xlabel('Longitude');
plt.ylabel('Latitude');
plt.axis('equal');
plt.grid(True);

plt.colorbar(label='Population (log$_{10}$ scale)');
plt.clim(2, 10);

# Generate legend for area sizes
area_range = [50, 100, 300, 500];
for area_size in area_range:
    plt.scatter([], [], c='k', alpha=0.4, s=area_size,
                label=str(area_size) + ' km$^2$');
plt.legend(labelspacing=0.8, title='City Area');


plt.show();
```
